```python
import pandas as pd

url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
```

# Reading Data

```python
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']
```

```python
df = pd.read_csv(url, names=column_names,
                          na_values='?', comment='\t',
                          sep=' ', skipinitialspace=True)
df
```

# Data Preprocessing

```python
df.isna().sum()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
columns = df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
df = Scaler.fit_transform(df)
df
```

```python
df = pd.DataFrame(df,columns=columns)
df
```

```python
X = df.drop(labels=["MPG"],axis=1)
X
```

```python
y = df[["MPG"]]
y
```

# Linear Regression Model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

# Polynomial Features


```python
from sklearn.preprocessing import PolynomialFeatures
poly_2 = PolynomialFeatures(degree=2 , include_bias=False)
x2 = poly_2.fit_transform(X)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(x2, y, test_size=0.3, random_state=42)
```

```python
model_2 = LinearRegression()
model_2.fit(X_train2,y_train2)
```

```python
model_2.score(X_train2,y_train2)
```

```python
model_2.score(X_test2,y_test2)
```

```python
poly_3 = PolynomialFeatures(degree=3 , include_bias=False)
x3 = poly_3.fit_transform(X)
```

```python
X_train3, X_test3, y_train3, y_test3 = train_test_split(x3, y, test_size=0.3, random_state=42)
```

```python
model_3 = LinearRegression()
model_3.fit(X_train3,y_train3)
```

```python
model_3.score(X_train3,y_train3)
```

```python
model_3.score(X_test3,y_test3)
```

```python
print(f"Third Polynomial Score = {model_3.score(X_test3,y_test3)}")
print(f"Second Polynomial Score = {model_2.score(X_test2,y_test2)}")
print(f"First Polynomial Score = {model.score(X_test,y_test)}")
```
